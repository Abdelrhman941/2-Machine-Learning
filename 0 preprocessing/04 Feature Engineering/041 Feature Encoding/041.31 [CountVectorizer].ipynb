```python
import pandas as pd                                                          # type: ignore
from sklearn.feature_extraction.text import CountVectorizer                  # type: ignore

vect = CountVectorizer()
# vect = CountVectorizer(stop_words=['call','you'])
# vect = CountVectorizer(lowercase= False)   # if ue want to keep the case of the words = False

simple_train = ['call you tonight', 'Call me a cab', 'please call me... PLEASE!']
## learn the 'vocabulary' of the training data (occurs in-place)
vect.fit(simple_train)

## examine the fitted vocabulary
words = vect.get_feature_names_out()
print(words)  
```

```python
# transform training data into a 'document-term matrix' 
simple_train_dtm = vect.transform(simple_train)
print(simple_train_dtm)   # >>>  (place in list , place in result filter)  repeat
```

```python
## convert sparse matrix to a usual "dense" matrix to understand
print(simple_train_dtm.toarray())

## or show it in dataframe 
pd.DataFrame(simple_train_dtm.toarray(), columns=vect.get_feature_names_out())
```

```python
## example text for model testing
simple_test = ["please don't call me"]

## transform testing data into a document-term matrix (using existing vocabulary)
simple_test_dtm = vect.transform(simple_test)
print(simple_test_dtm.toarray())

## examine the vocabulary and document-term matrix together
pd.DataFrame(simple_test_dtm.toarray(), columns=vect.get_feature_names_out())
```
