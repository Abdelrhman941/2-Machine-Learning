```python
from sklearn.datasets import load_breast_cancer     # type: ignore

X , y = load_breast_cancer(return_X_y=True)
```

```python
# Splitting data:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)
print(f"X_train shape: {X_train.shape}")
print(f"X_test  shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test  shape: {y_test.shape}")
```

```python
#Applying Ridge Regression Model 
RidgeRegressionModel = Ridge(alpha=1.0,random_state=33)
RidgeRegressionModel.fit(X_train, y_train)

## Calculating Details:
print(f'Ridge Regression Train Score : {RidgeRegressionModel.score(X_train, y_train)}')
print(f'Ridge Regression Test  Score : {RidgeRegressionModel.score(X_test, y_test)}\n')
print(f'Ridge Regression Coef : \n{RidgeRegressionModel.coef_}\n')
print(f'Ridge Regression intercept : {RidgeRegressionModel.intercept_}')
```

```python
## Calculating Prediction:
y_pred = RidgeRegressionModel.predict(X_test)
print(f'Predicted Value for Ridge Regression : \n{ y_pred[:10]}\n')
print(f'actual    Value for Ridge Regression : \n{ y_test[:10]}')
```

```python
# Calculating Mean Absolute Error:
MAEValue = mean_absolute_error(y_test, y_pred, multioutput='uniform_average')   # it can be raw_values
print(f'Mean   Absolute Error Value : {MAEValue}')
##########################################################
#Calculating Mean Squared Error
MSEValue = mean_squared_error(y_test, y_pred, multioutput='uniform_average')    # it can be raw_values
print(f'Mean   Squared  Error Value : {MSEValue}')
##########################################################
#Calculating Median Squared Error
MdSEValue = median_absolute_error(y_test, y_pred)
print(f'Median Squared  Error Value : {MdSEValue}')
```
