```python
from sklearn import datasets, svm       # type: ignore
import matplotlib.pyplot as plt         # type: ignore

digits = datasets.load_digits()
X = digits.data
y = digits.target

x_train = X[:-100]
x_test = X[-100:]
y_train = y[:-100]
y_test =  y[-100:]
```

```python
svc = svm.SVC(C=1, kernel='linear')
svc.fit(x_train,y_train )
sc = svc.score(x_test,y_test)
print(f'score = {sc}')
```

```python
for j in range( 1,500,20):
    print(f'Prediction >> {svc.predict( X[j].reshape(1,-1) )}')
    plt.imshow(digits.images[j],cmap=plt.cm.gray_r, interpolation='nearest')
    plt.show()
```
