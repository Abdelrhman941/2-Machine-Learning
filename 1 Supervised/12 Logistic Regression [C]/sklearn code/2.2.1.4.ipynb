```python
import numpy as np                                                      # type: ignore
from sklearn.datasets import load_iris                                  # type: ignore
from sklearn.metrics import accuracy_score , confusion_matrix           # type: ignore
from sklearn.model_selection import StratifiedKFold                     # type: ignore
from sklearn.linear_model import LogisticRegression                     # type: ignore

X, y = load_iris(return_X_y=True)
```

```python
skf = StratifiedKFold(n_splits = 10)
predict = np.zeros(y.shape[0])
```

```python
for train,test in skf.split(X,y):
    x_train = X[train]
    x_test = X[test]
    y_train = y[train]
    y_test = y[test]
    logreg = LogisticRegression()
    logreg.fit(x_train , y_train)
    result = logreg.predict(x_test)
    predict[test] = result
    
    print(f'train data :\n{train}')
    print(f'test data :\n{test}')
    print(f'result = {result}')
    print(f'this accuracy = {accuracy_score(y_test , result):.2f}')
    print('===============================\n')
```

```python
print(f'total accuracy   : {accuracy_score(y , predict):.2f}')

conf = confusion_matrix(y , predict)
print(f'confusion matrix :\n{conf}')
```
